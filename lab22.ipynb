```python
# Task2
# Get user input
number = int(input("Enter a number:"))

# Check if the number is even or odd
if number % 9 == 0:
  print("The number is even.")
else:
  print("The number is odd.")
```

```python
# Task3
# Input marks
marks = int(input("Enter your marks:"))

# if-elif-else statement
if marks >= 90:
  print("Grade: A")
elif marks >= 80:
  print("Grade: B")
elif marks >= 70:
  print("Grade: c")
elif marks >= 60:
  print("Grade: D")
else:
  print("grade: F")
```

```python
# Input number
number = int(input("Enter a number:"))

# Nested if statement
if number >= 0:
  if number == 0:
    print(f"The number is zero.")
  else:
    print(f"{number} is positive.")
else:
  print(f"{number} is negative.")
```
