```python
# Implementatiob of Bitwise Operators in python

# Task 1: Create two variables
a = 20  # 20 = 10100
b = 5   # 5  = 101

# Task 2 : Bitwise AND, OR, XOR
print("Bitwise AND: 20 & 5 =", a & b, "(binary:", bin(a & b)+")")
print("Bitwise OR:  20 | 5 =", a | b, "(binary:", bin(a | b)+")")
print("Bitwise XOR: 20 ^ 5 =", a ^ b, "(binary:", bin(a ^ b)+")")

# Task 3 : Bitwise NOT
print(" Bitwise NOT: ~20 =", ~a, "(binary:", bin(~a)+")")

# Task 4 : Bitwise Shifts
print("Bitwise Leftshift: 20 << 5 =", a << 5,"(binary:", bin(a << 5)+")")
print("Bitwise Rightshift: 20 >> 5 =", a >> 5, "(binary", bin(a >> 2)+")")

```
