```python
# Define two variables
a = 30
b = 40

# Equal to
print(f"{a} == {b} : {a == b}")

# Not equal to
print(f"{a} != {b} : {a != b}")

# Greater than
print(f"{a} > {b} : {a > b}")

# Less than
print(f"{a} < {b} : {a < b }")

# Greater than or equal to
print(f"{a} >= {b} : {a >= b}")

# Less than or equal to
print(f"{a} <= {b} : {a <=b}")

# Additional example with strings
str1 = " pranavi"
str2 = "amulya"

# String comparisons
print(f"'{str1}' == '{str2}' : {str1 == str2}")
print(f"'{str1}' != '{str2}' : {str1 != str2}")
print(f"'{str1}' > '{str2}' : {str1 > str2}")
print(f"'{str1}' < '{str2} : {str1 < str2}")
```
